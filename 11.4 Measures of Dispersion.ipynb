# **Measures of Dispersion**

**To understand how spread out or varied your data is. Measures of central tendency tell us the center, but dispersion tells us how far the data points are from that center.**

## **Types of Measures of Dispersion:**

| Measure                 | What it Tells                                    | Best Use                        |
| ----------------------- | ------------------------------------------------ | ------------------------------- |
| **Range**               | Max - Min                                        | Simple overview                 |
| **Mean Deviation (MD)** | Average of absolute differences from mean/median | When you want average variation |
| **Variance**            | Average squared deviation from mean              | Step before Standard Deviation  |
| **Standard Deviation**  | Square root of Variance                          | Most common measure             |
| **Interquartile Range** | Q3 - Q1                                          | Focus on middle 50% (robust)    |


### **Range**

**The difference between the maximum and minimum values.**

**Formula:**

Range=Max(x)âˆ’Min(x)

```python
data = [70, 85, 78, 85, 90, 88, 85, 75, 95, 85]
range_value = max(data) - min(data)
print("Range:", range_value)
```

### **Mean Deviation**

**The average of the absolute differences between each value and the mean (or median).**

```python
import numpy as np
data = np.array([70, 85, 78, 85, 90, 88, 85, 75, 95, 85])

mean_md = np.mean(np.abs(data - np.mean(data)))
median_md = np.mean(np.abs(data - np.median(data)))

print("Mean Deviation (about Mean):", mean_md)
print("Mean Deviation (about Median):", median_md)
```

### **Variance**

**The average of the squared differences from the mean.**

```python
variance = np.var(data)  # population variance
print("Variance:", variance)
```

### **Standard Deviation (SD)**

**The square root of variance. It shows how much the values deviate from the mean in original units.**

```python
std_dev = np.std(data)
print("Standard Deviation:", std_dev)
```

### **Interquartile Range (IQR)**

**Difference between Q3 (75th percentile) and Q1 (25th percentile). It ignores outliers and focuses on the middle 50%.**

**Formula:**

IQR=Q3âˆ’Q1

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Interquartile Range (IQR):", IQR)
```

# **Full Python Code**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = np.array([70, 85, 78, 85, 90, 88, 85, 75, 95, 85])
df = pd.DataFrame({'Marks': data})

# -------------------------------
# ðŸ“¦ Box Plot
# -------------------------------
plt.figure(figsize=(8, 4))
plt.boxplot(df['Marks'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Marks')
plt.xlabel('Marks')
plt.grid(True)
plt.show()

# -------------------------------
# ðŸ“‰ Histogram with Standard Deviation
# -------------------------------
mean_val = df['Marks'].mean()
std_dev = df['Marks'].std()

plt.figure(figsize=(8, 5))
plt.hist(df['Marks'], bins=7, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean = {mean_val:.1f}')
plt.axvline(mean_val + std_dev, color='blue', linestyle='--', label=f'Mean + SD')
plt.axvline(mean_val - std_dev, color='blue', linestyle='--', label=f'Mean - SD')
plt.title('Histogram of Marks with Mean Â± SD')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

data = np.array([70, 85, 78, 85, 90, 88, 85, 75, 95, 85])

# Range
range_val = np.max(data) - np.min(data)

# Mean Deviation
mean_md = np.mean(np.abs(data - np.mean(data)))
median_md = np.mean(np.abs(data - np.median(data)))

# Variance & Standard Deviation
variance = np.var(data)  # population
std_dev = np.std(data)

# Interquartile Range
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
iqr = Q3 - Q1

# Print Results
print("Range:", range_val)
print("Mean Deviation (Mean):", mean_md)
print("Mean Deviation (Median):", median_md)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Interquartile Range (IQR):", iqr)
```
