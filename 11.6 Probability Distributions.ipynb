# **Probability Distributions**

**A probability distribution describes how probabilities are assigned to all possible outcomes of a random variable**

**A random variable is a quantity that can take on different values based on the outcome of a random event (like tossing a coin or rolling a die).**

## **Types of Probability Distributions**

1**Ô∏èDiscrete Distribution**
    
**A probability distribution where the random variable takes countable values, like 0, 1, 2, etc.**

üëâ Example: Number of heads in coin tosses.

2Ô∏è‚É£**Continuous Distribution**
    
**A distribution where the random variable takes infinite values within a range, like 1.2, 1.25, 1.35, etc.**

üëâ Example: Height, weight, temperature.

## **Binomial Distribution**

**A discrete distribution used for fixed number of trials, each with 2 outcomes (success/failure), and constant probability.**

üëâ Example: Probability of getting 3 heads in 5 coin tosses.

**Used when:**

A fixed number of trials (n)

Only 2 outcomes (success/failure)

Trials are independent

Each trial has same probability of success (p)

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 5
p = 0.5
x = range(0, n + 1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color='skyblue')
plt.title('Binomial Distribution (n=5, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

## **Poisson Distribution**

**üìñ Definition:**

**Used for modeling the number of events in a fixed interval (time, area, volume), especially when events are rare and occur independently.**

**A discrete distribution used to model the number of rare events occurring in a fixed time or space, where events happen independently.**

üëâ Example: Number of emails received per hour.

```python
from scipy.stats import poisson

lam = 4
x = range(0, 10)
pmf = poisson.pmf(x, mu=lam)

plt.bar(x, pmf, color='lightgreen')
plt.title('Poisson Distribution (Œª=4)')
plt.xlabel('Number of Requests')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

## **Normal Distribution (Bell Curve)**

**üìñ Definition:**

**Used to model continuous data that is symmetrical and bell-shaped (most values near the mean).**

```python
from scipy.stats import norm
import numpy as np

mu = 70
sigma = 10
x = np.linspace(40, 100, 100)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='blue')
plt.title('Normal Distribution (Œº=70, œÉ=10)')
plt.xlabel('Score')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

##  **Properties of Normal Distribution**

Symmetrical about the mean, Œº


Mean = Median = Mode

Total area under curve = 1

Empirical Rule (68-95-99.7 Rule):

| Range             | Coverage |
| ----------------- | -------- |
| $\mu \pm 1\sigma$ | 68%      |
| $\mu \pm 2\sigma$ | 95%      |
| $\mu \pm 3\sigma$ | 99.7%    |


## **Z-Scores and Standard Normal Distribution**

**üìñ Definition:**

**A Z-score tells how many standard deviations a value is from the mean.**

```python
# Z-score for X = 85
X = 85
mu = 70
sigma = 10
z = (X - mu) / sigma
print("Z-score:", z)

# Probability of scoring less than 85
prob = norm.cdf(z)
print("P(X < 85):", round(prob, 4))
```
