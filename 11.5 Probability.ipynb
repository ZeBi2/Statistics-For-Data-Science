# **Probability**

## **What is Probability?**

Probability is the measure of the likelihood that an event will occur.
It always lies between 0 and 1:

0 → Impossible event

1 → Certain event


 


## **Basic Probability Terms:**

| Term                  | Meaning                                          |
| --------------------- | ------------------------------------------------ |
| **Experiment**        | An action like rolling a die                     |
| **Sample Space**      | All possible outcomes (e.g., {1, 2, 3, 4, 5, 6}) |
| **Event**             | One or more outcomes (e.g., rolling even number) |
| **Favorable Outcome** | The result you're interested in                  |


## **Types of Events:**

### **Independent Events:**

Events where the outcome of one does not affect the other
(e.g., tossing 2 coins)

### **Dependent Events:**

Events where the outcome of one affects the other
(e.g., drawing 2 cards without replacement)

```python
P_D = 0.01
P_not_D = 0.99
P_pos_given_D = 0.95
P_pos_given_not_D = 0.05

P_D_given_pos = (P_pos_given_D * P_D) / (
    P_pos_given_D * P_D + P_pos_given_not_D * P_not_D
)
print("P(Disease | Positive Test):", round(P_D_given_pos, 4))
```
