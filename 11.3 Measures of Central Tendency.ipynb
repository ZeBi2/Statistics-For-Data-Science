# **Measures of Central Tendency**

# **Definition:**

**Measures of Central Tendency are statistical values that describe the center or average of a dataset.**.

## **Three Main Types:**

| Measure    | Meaning                          | Best Used When                 |
| ---------- | -------------------------------- | ------------------------------ |
| **Mean**   | Arithmetic average               | Data is evenly distributed     |
| **Median** | Middle value when data is sorted | Data has outliers or is skewed |
| **Mode**   | Most frequently occurring value  | You want the most common value |

## **Python Example**

```python
import pandas as pd
import numpy as np
from scipy import stats

# Sample dataset
data = {
    'Marks': [70, 85, 78, 85, 90, 88, 85, 75, 95, 85]
}
df = pd.DataFrame(data)

# Mean
mean_value = df['Marks'].mean()

# Median
median_value = df['Marks'].median()

# Mode
mode_value = df['Marks'].mode()[0]  # mode() returns a Series

# Print results
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Marks': [70, 85, 78, 85, 90, 88, 85, 75, 95, 85]
}
df = pd.DataFrame(data)

# Calculate measures
mean_value = df['Marks'].mean()
median_value = df['Marks'].median()
mode_value = df['Marks'].mode()[0]

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(df['Marks'], bins=7, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_value:.1f}')
plt.axvline(median_value, color='green', linestyle='--', linewidth=2, label=f'Median = {median_value:.1f}')
plt.axvline(mode_value, color='purple', linestyle='--', linewidth=2, label=f'Mode = {mode_value:.1f}')

# Title and labels
plt.title('Distribution of Marks with Mean, Median, and Mode')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
